```python
# importing the dependencies
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
data_frame = pd.DataFrame(dataset.data,columns = dataset.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label'] = dataset.target
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.isnull().sum()
```

```python
data_frame.drop(columns='label',axis=1)
```

```python
#1 --> Benign
#0 --> Malignant
```

```python
x = data_frame.drop(columns='label',axis=1)
y = data_frame['label']
```

```python
print(x)
print(y)
```

```python
x = np.asarray(x)
y = np.asarray(y)
```

***GridSearchCV***

```python
model = SVC()
```

```python
parameters = {
              'kernel':['linear','poly','rbf','sigmoid'],
              'C':[1, 5, 10, 20]
}
```

```python
# grid search
classifier = GridSearchCV(model, parameters, cv=6)
```

```python
# fitting the data to our model
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
highest_acc = classifier.best_score_
print(highest_acc)
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```
